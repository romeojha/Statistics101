```python
import random
age=[]
for i in range(300000):
    age.append(random.randint(1,70))
```

```python
import seaborn as sns
sns.distplot(age)
```

```python
import pandas as pd
age=pd.DataFrame(age)
```

```python
sample=[]
for i in range(10000):
    sample.append(age.sample(n=30).mean())
```

```python
sns.distplot(sample)
```

```python
sample1=pd.Series(sample) 
```

```python
x=(sample1-sample1.mean())/sample1.std()
```

```python
sns.distplot(x)
```

```python
import seaborn as sns
df=sns.load_dataset("iris")
import numpy as np
```

```python
df.head()
```

```python
sns.kdeplot(data=df,x=df["sepal_length"])
```

```python
temp=sorted(df["sepal_length"].tolist())
```

```python
y_quant=[]
for i in range(1,101):
    y_quant.append(np.percentile(temp,i))
```

```python
sample=np.random.normal(loc=0,scale=1,size=1000)
```

```python
x_quant=[]
for i in range(1,101):
    x_quant.append(np.percentile(sample,i))
```

```python
sns.scatterplot(x=x_quant,y=y_quant)
```

```python
#using statsmodel to qq plot the 
import statsmodels.api as sm
sm.qqplot(df["sepal_length"],line='45',fit=True)
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

sample=np.random.uniform(high=1,low=0,size=1000)
uni=np.random.uniform(high=1,low=0,size=1000)

```

```python
params=stats.uniform.fit(uni)
dist=stats.uniform(loc=params[0],scale=params[1])
print(params)
```

```python
fig=sm.qqplot(sample,dist=dist,line='45')
plt.title("qq plot")
plt.xlabel("theoritical quantiles")
plt.ylabel('sample quantiles')
```

```python
sm.qqplot(sample,line="45",fit=True)
```
