```python
import pandas as pd
```

```python
import random
l=[]
for i in range(10000):
    l.append(random.randint(1,6))# rolling a dice 10000 times 
```

```python
l[:5]
```

```python
d=(pd.Series.value_counts(l)/pd.Series.value_counts(l).sum()).sort_index(ascending=True)
```

```python
d
```

```python
import seaborn as sns
sns.barplot(x=d.index,y=d.values)
```

calculating cumulative distribution 

```python
# for cumulative distribution function

cumsum=pd.Series.cumsum(d)
```

```python
cumsum
```

```python
sns.barplot(x=cumsum.index,y=cumsum.values)
```

calculating cumulative distribution for three dice and chance to get respective counts
 

```python
three_dice_count=[]
for i in range(10000):
    a=random.randint(1,6)
    b=random.randint(1,6)
    c=random.randint(1,6)
    three_dice_count.append(a+b+c)
```

```python
three_dice_count[:5]
```

```python
d=(pd.Series.value_counts(three_dice_count)/pd.Series.value_counts(three_dice_count).sum()).sort_index(ascending=True)
```

```python
d
```

```python
sns.barplot(x=d.index,y=d.values)
```

```python
cumsum=pd.Series.cumsum(d)
```

```python
sns.barplot(x=cumsum.index,y=cumsum.values)
```
